```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

```

```python
def run_main():
    img = cv2.imread('IMG_8253.JPG',0)
    cimg = cv2.cvtColor(img,cv2.COLOR_GRAY2BGR)
    img = cv2.medianBlur(img,55)
    
    circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,0.9,120,param1=50,param2=30,minRadius=100,maxRadius=190)

    circles = np.uint16(np.around(circles))
    for i in circles[0,:]:
        cv2.circle(cimg,(i[0],i[1]),i[2],(0,255,255),18)
        
    plt.imshow(cimg)
    
if __name__ == "__main__":
    run_main()
```

```python
def run_main():
    img = cv2.imread('IMG_8256.JPG',0)
    cimg = cv2.cvtColor(img,cv2.COLOR_GRAY2BGR)
    img = cv2.bilateralFilter(img,25,100,100)
    img = cv2.medianBlur(img,5)
    
    circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,0.9,120,param1=50,param2=30,minRadius=100,maxRadius=190)

    circles = np.uint16(np.around(circles))
    for i in circles[0,:]:
        cv2.circle(cimg,(i[0],i[1]),i[2],(0,255,255),18)

    plt.imshow(cimg)

if __name__ == "__main__":
    run_main()
```
