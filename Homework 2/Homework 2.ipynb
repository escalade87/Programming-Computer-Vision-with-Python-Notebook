```python
# import the necessary packages
import numpy as np
import cv2
from matplotlib import pyplot as plt

# construct a sharpening filter
sharpen = np.array((
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]), dtype="int")

# construct the Edge Detection A
edgeA = np.array((
    [1, 0, -1],
    [0, 0, 0],
    [-1, 0, 1]), dtype="int")

# construct the Edge Detection B
edgeB = np.array((
    [0, 1, 0],
    [1, -4, 1],
    [0, 1, 0]), dtype="int")

# construct the Edge Detection C
edgeC = np.array((
    [-1, -1, -1],
    [-1, 8, -1],
    [-1, -1, -1]), dtype="int")

# construct the Box Blur
boxBlur = np.ones((3, 3), dtype="float") * (1.0 / 9.0)


# construct the Gaussian Blur 3x3
gaussianBlur3 = np.array((
    [1, 2, 1],
    [2, 4, 2],
    [1, 2, 1]), dtype="float") * (1.0 / 16.0)

# construct the Gaussian Blur 5x5
gaussianBlur5 = np.array((
    [1, 4, 6, 4, 1],
    [4, 16, 24, 16, 4],
    [6, 24, 36, 24, 6],
    [4, 16, 24, 16, 4],
    [1, 4, 6, 4, 1]), dtype="float") * (1.0 / 256.0)

# construct the kernel bank, a list of kernels we're going
# to apply using both our custom `convole` function and
# OpenCV's `filter2D` function
kernelBank = (
    ("Sharpen", sharpen),
    ("Edge Detection A", edgeA),
    ("Edge Detection B", edgeB),
    ("Edge Detection C", edgeC),
    ("Box Blur", boxBlur),
    ("Gaussian Blur 3x3", gaussianBlur3),
    ("Gaussian Blur 5x5", gaussianBlur5)
)

# load the input image and convert it to grayscale
image = cv2.imread("BM1.JPG")
gray = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

print("[INFO] Original Image")
plt.imshow(gray)
plt.axis('off')
plt.show()

# loop over the kernels
for (kernelName, kernel) in kernelBank:
    # apply the kernel to the grayscale image using both
    # our custom `convole` function and OpenCV's `filter2D`
    # function
    print("[INFO] applying {} kernel".format(kernelName))
    opencvOutput = cv2.filter2D(gray, -1, kernel)
    
    # show the output images
    plt.imshow(opencvOutput)
    plt.axis('off')
    plt.show()
```
