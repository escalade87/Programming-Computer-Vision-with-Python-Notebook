```python
from PIL import Image
from pylab import *
import matplotlib.pyplot as pl
# read image to array
im = array(Image.open('data/turningtorso1.jpg'))
# plot the image
imshow(im)
# some points
x = [100,100,300,300]
y = [100,400,100,400]
# plot the points with red star-markers
plot(x,y,'ks:')
# line plot connecting the first two points
plot(x[:2],y[:2])
# add title and show the plot
title('Plotting: "turningtorso1.jpg"')
show()
#plot(x,y) # default blue solid line
#plot(x,y,’r*’) # red star-markers
#plot(x,y,’go-’) # green line with circle-markers
#plot(x,y,’ks:’) # black dotted line with square-markers
```

```python
# read image to array
im = array(Image.open('data/turningtorso1.jpg').convert('L'))
# create a new figure
figure()
# don’t use colors
gray()
imshow(im)
figure()
# show contours with origin upper left corner
contour(im, origin='image')
axis('equal')
axis('off')
show()
```
